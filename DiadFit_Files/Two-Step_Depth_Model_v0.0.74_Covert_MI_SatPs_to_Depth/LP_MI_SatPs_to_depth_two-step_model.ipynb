Converting Melt Inclsusion Pressures to Depth Using two-step density model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import DiadFit as pf
pf.__version__
```

Loading in MI Saturation Pressures

```python
df=pd.read_excel('LP_MI_VESIcal_multimodelSatPs_for_twostepdepth.xlsx')
df
```

```python
#measured MI CO2
Depth1=pf.convert_pressure_to_depth(P_kbar=df['SaturationP_bars_VESIcal_MagmaSat_Actual']/1000, model='two-step', d1=14, rho1=2800, rho2=3100)
#Depth1.to_excel('SaturationP_bars_VESIcal_MagmaSat_Actual_output_two_step.xlsx', index=True) 
Depth1
```

```python
#plus 1sigma
Depth_high=pf.convert_pressure_to_depth(P_kbar=df['SaturationP_bars_VESIcal_MagmaSat_plus1sigma']/1000, model='two-step', d1=14, rho1=2800, rho2=3100)
#Depth_high.to_excel('SaturationP_bars_VESIcal_MagmaSat_plus1sigma_output_two_step.xlsx', index=False) 
Depth_high
```

```python
#minus 1sigma
Depth_low=pf.convert_pressure_to_depth(P_kbar=df['SaturationP_bars_VESIcal_MagmaSat_minus1sigma']/1000, model='two-step', d1=14, rho1=2800, rho2=3100)
#Depth_low.to_excel('SaturationP_bars_VESIcal_MagmaSat_minus1sigma_output_two_step.xlsx', index=False) 
Depth_low
```
